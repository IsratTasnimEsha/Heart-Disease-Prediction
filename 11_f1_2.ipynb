```python
import pandas as pd
import os
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, roc_curve
os.chdir('Resources/')

import warnings
warnings.filterwarnings("ignore")

df = pd.read_csv('10_SP_Preprocessed_Data.csv')
X = df.drop(['HeartDisease'], axis='columns')
Y = df[['HeartDisease']]
```

```python
from xgboost import XGBClassifier

mx=0
ind=0
for i in range (5000):
    seed=i

    full = []
    
    for j in range(1, 10):
        X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1*j, random_state=seed)
        
        xgb = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=seed)

        model_xgb = xgb.fit(X_train, Y_train)
        
        pred_xgb = model_xgb.predict(X_test)

        accuracy_xgb = f1_score(Y_test, pred_xgb)
    
        full.append(accuracy_xgb)

    accuracy_nb = sum(full)
    if(accuracy_nb>mx):
        mx=accuracy_nb
        ind=i
    
    print(i, accuracy_nb, mx, ind)
```
