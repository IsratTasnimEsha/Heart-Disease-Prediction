```python
import os
import csv
import pandas as pd
import numpy as np
import pickle
import random
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder
from math import ceil
```

```python
def power(a, b, c):
    x = 1
    y = a

    while b > 0:
        if b % 2 != 0:
            x = (x * y) % c
        y = (y * y) % c
        b = int(b / 2)

    return x % c
```

```python
public_key_file_path = 'Public_Key.pkl'
with open(public_key_file_path, 'rb') as public_key_file:
    public_key = pickle.load(public_key_file)
q, h, g = public_key
```

```python
input_file_path = 'Encoded_Data.csv'
df = pd.read_csv(input_file_path)
```

```python
def encrypt_value(value, q, h, g, k):
    s = power(h, k, q)
    p = power(g, k, q)
    encrypted_value = (p, s * value % q)
    return encrypted_value
```

```python
k_values = {}
encrypted_data = {}
for column in df.columns:
    encrypted_column = []
    if column not in k_values:
        k_values[column] = random.randint(2, q-1)
    k = k_values[column]
    for value in df[column]:
        encrypted_value = encrypt_value(value, q, h, g, k)
        encrypted_column.append(encrypted_value)
    encrypted_data[column] = encrypted_column
```

```python
def symmetric_encrypt(value):
    hashed_value = hash(value)
    encrypted_value = hashed_value % 1000000 
    return encrypted_value
```

```python
def symmetric_decrypt(encrypted_value):
    return encrypted_value

os.chdir('..')
with open('4_Encrypted_Dataset\Decryption_Key.pkl', 'wb') as key_file:
    pickle.dump(symmetric_decrypt, key_file)
os.chdir(os.getcwd())
```

```python
encrypted_df = pd.DataFrame(encrypted_data)
encrypted_df = encrypted_df.applymap(symmetric_encrypt)

encrypted_df.to_csv('4_Encrypted_Dataset\Encrypted_Data.csv', index=False)
```
