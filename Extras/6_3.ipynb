```python
import pandas as pd
import os
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, roc_curve
os.chdir('Resources/')

df = pd.read_csv('5_Preprocessed_Data.csv')
X = df.drop(['HeartDisease'], axis='columns')
Y = df[['HeartDisease']]
```

```python
from sklearn.ensemble import RandomForestClassifier

mx=0
ind=0
for i in range (64):
    seed=i
    rf = RandomForestClassifier(random_state=seed)

    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, random_state=seed)

    model_rf = rf.fit(X_train, Y_train)

    pred_rf = model_rf.predict(X_test)

    accuracy_rf = roc_auc_score(Y_test, pred_rf)

    if(accuracy_rf>mx):
        mx=accuracy_rf
        ind=i
    print(i, accuracy_rf, mx, ind)
```
