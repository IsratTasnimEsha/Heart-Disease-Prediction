```python
import os
import csv
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
os.chdir('Resources/')
```

```python
df = pd.read_csv('1_Structured_Data.csv')
df.head()
```

```python
age_bins = [20, 30, 40, 50, 60, 70, 80, 90, 100]
age_labels = ['20-29', '30-39', '40-49', '50-59', '60-69', '70-79', '80-89', '90-99']

df['AgeGroup'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels, right=False)

plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")

ax = sns.countplot(x='AgeGroup', hue='HeartDisease', data=df, palette='coolwarm', order=age_labels)

plt.title('Distribution of Heart Disease by Age Group', fontsize=16, fontweight='bold', family='serif')
plt.xlabel('Age Group', fontsize=14, fontweight='bold', family='serif', labelpad=10)
plt.ylabel('Count', fontsize=14, fontweight='bold', family='serif', labelpad=10)

plt.xticks(fontsize=12, fontweight='bold', family='serif')
plt.yticks(fontsize=12, fontweight='bold', family='serif')

plt.legend(title='Heart Disease', fontsize=12, title_fontsize=12, shadow=True)

sns.despine(left=True, bottom=True)

plt.show()
```
