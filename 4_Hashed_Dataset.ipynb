```python
import pandas as pd
import hashlib
import os
os.chdir('Resources/')
```

```python
def hash_value(value):
    hash_object = hashlib.sha256(str(value).encode())
    hash_hex = hash_object.hexdigest()
    hash_int = int(hash_hex, 16) % (2**32)
    return hash_int
```

```python
df = pd.read_csv('3_Encrypted_Data.csv')
X = df.drop(['HeartDisease'], axis='columns')
Y = df[['HeartDisease']]
```

```python
hashed_X = X.applymap(hash_value)
hashed_df = pd.concat([hashed_X, Y], axis=1)

hashed_df.to_csv('4_Hashed_Data.csv', index=False)
```
